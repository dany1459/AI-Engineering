```python
#DO THE NECESSARY IMPORTS
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# loading the data
df = pd.read_csv(r'C:\Users\Igor\Documents\GitHub\AI-Engineering\Chapter 2\01. Intro to ML\Part 1\data\diabetes.csv')
# set x and y for features and labels
y = df['Outcome']
X = df.drop('Outcome', axis=1)
```

```python
#define the class
class LogisticRegression:

    # initialize the constructor method
    def __init__(self, lr=0.01, num_iter=100000, fit_intercept=True, verbose=False):
        self.lr = lr
        self.num_iter = num_iter
        self.fit_intercept = fit_intercept
        self.verbose = verbose
    
    # z = wx + b
    # b bias
    # 3 private methods
    def __add_intercept(self, X):
        intercept = np.ones((X.shape[0], 1))
        return np.concatenate((intercept, X), axis=1)
    
    # define sigmoid function
    def __sigmoid(self, z):
        return 1 / (1 + np.exp(-z))

    # define loss function
    def __loss(self, h, y):
        return (-y * np.log(h) - (1 - y) * np.log(1 - h)).mean()
    

    def fit(self, X, y):
        if self.fit_intercept:
            X = self.__add_intercept(X)
        
        # w weights initialization
        self.theta = np.zeros(X.shape[1])
        
        for i in range(self.num_iter):
            z = np.dot(X, self.theta)
            h = self.__sigmoid(z)
            # calculate the gradient decent to see how loss changes with different params
            gradient = np.dot(X.T, (h - y)) / y.size
            # update the weights with the learning rate
            self.theta -= self.lr * gradient
            
            if(self.verbose == True and i % 10000 == 0):
                z = np.dot(X, self.theta)
                h = self.__sigmoid(z)
                print(f'loss: {self.__loss(h, y)} \t')
    
    # predict the probability that X belongs to class 1 or 0
    def predict_prob(self, X):
        if self.fit_intercept:
            X = self.__add_intercept(X)
    
        return self.__sigmoid(np.dot(X, self.theta))
    
    def predict(self, X, threshold):
        return self.predict_prob(X) >= threshold
```

```python
model = LogisticRegression(lr=0.1, num_iter=300000)
model.fit(X, y)
preds = model.predict_prob(X)
preds
```

```python
preds = model.predict(X, 0.1)
# accuracy
(preds == y).mean()
```
