### Exercise 1: Read the Total profit of all months and show it using a line plot.
Total profit data provided for each month. The enerated line plot must include the following properties:<br>
- X label name = Month Number
- Y label name = Total profit

The line plot graph should look like this:


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Write your code here
url = './company_sales_data.csv'
data = pd.read_csv(url)
df = pd.DataFrame(data)
df.plot.line(x='month_number', y='total_profit')
```

### Exercise 2: Get the Total profit of all months and show a line plot with the following Style properties:
- Line Style dotted and Line color should be red
- Show legend at the lower right location.
- X label name = Month Number
- Y label name = Sold units number
- Add a circle marker.
- Line marker color as read
- Line width should be 3

The line plot graph should look like this:


```python
df.plot.line(x='month_number', y='total_profit', marker='o', markerfacecolor='black' ,c='red', linestyle='dotted', linewidth=3)
plt.legend(loc='lower right')
plt.xlabel('Month Number')
plt.ylabel('Sold Units Number')

```

### Exercise 3: Read all product sales data and show it using a multiline plot.
Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product).

The graph should look like this:


```python
data.head()
```

```python
plt.plot(df['month_number'], df['facecream'], label='Face cream sales')
plt.plot(df['month_number'], df['facewash'], label='Facewash sales')
plt.plot(df['month_number'], df['toothpaste'], label='Toothpaste Sales')
plt.plot(df['month_number'], df['bathingsoap'], label='Bathing Soap Sales')
plt.plot(df['month_number'], df['shampoo'], label='Shampoo Sales')
plt.plot(df['month_number'], df['moisturizer'], label='Moisturizer Sales')

plt.style.use('bmh')
plt.legend(loc=(1, 0.5))
plt.xlabel('Month Number')
plt.ylabel('Number of Units Sold')
plt.yticks([0, 1000, 2000, 4000, 6000, 8000, 10000, 12500, 15000])
plt.show()
```

### Exercise 4: Read the toothpaste sales data of each month and show it using a scatter plot.
Also, add a grid in the plot (the grid line style should be âââ).

The scatter plot should look like this:


```python
df.plot.scatter(x='month_number', y='toothpaste', label='Toothpaste sales')

plt.xlabel('Month Number')
plt.ylabel('Units of Toothpaste Sold')
plt.legend(loc='best')
plt.grid(True, linewidth=0.5, linestyle='--')

plt.show()
```

###Â Exercise 5: Read the face cream and facewash product sales data and show it using the bar chart.
The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

The bar chart should look like this:


```python
months = df['month_number'].tolist()

plt.bar(months, df['facecream'], width=0.3, label='Face Cream sales', align='edge')
plt.bar(months, df['facewash'], width=-0.3, label='Facewash sales', align='edge')

plt.legend(loc='best')
plt.xticks(months)

plt.show()
```

### Exercise 6: Read the sales data of bathing soap of all months and show it using a bar chart. Save this plot to your hard disk.
The bar chart should look like this:


```python
plt.bar(months, df['bathingsoap'], width=0.5, label='Bathing Soap sales')

plt.xticks(months)
plt.legend(loc='best')
plt.grid(True, linewidth=0.5, linestyle='--')
plt.title('Bathing Soap Sales')

plt.savefig('D:\stuff\coding\_bathing_soap_sales.jpg')
plt.show()
```

### Exercise 8: Calculate the total sale data for last year for each product and show it using a Pie chart.
Note: In the Pie chart you should display the Number of units sold per year for each product in percentage.

The Pie chart should look like this:


```python
total_sales = [ df['facecream'].sum(), df['facewash'].sum(), df['toothpaste'].sum(), df['bathingsoap'].sum(), df['shampoo'].sum(), df['moisturizer'].sum() ]

plt.pie(total_sales, labels=['Face Cream', 'Facewash', 'Toothpaste', 'Bathing Soap', 'Shampoo', 'Moisturizer'], autopct='%1.1f%%')

plt.legend(loc=(1.25, 0.5))

plt.show()
```

###Â Exercise 9: Read the Bathing soap facewash of all months and display it using the Subplot.
The Subplot should look like this:


```python
fig,ax = plt.subplots(2, sharex=True)

ax[0].plot(months, df['bathingsoap'], color='black', marker='o', linewidth=2)
ax[0].set_title('Sales data of Bathing Soap')

ax[1].plot(months, df['facewash'], color='red', marker='o', linewidth=2)
ax[1].set_title('Sales data of Facewash')

plt.xticks(months)
plt.xlabel('Month Number')
plt.ylabel('Sales units in numbers')

plt.show()
```
