```python
# Strive2
import numpy as np
import matplotlib.pyplot as plt
import cv2
cv2.__version__
```

```python
from yolo_helpers import *
```

```python
models_configuration_path = 'models/yolov3.cfg'
models_weights_path = 'models/yolov3.weights'
```

```python
net = cv2.dnn.readNetFromDarknet(models_configuration_path, models_weights_path)
net
```

```python
img = cv2.imread('images/dog.jpg')
plt.imshow(img)
```

```python
# create a blob from the frame or image
# arg -> swapRB=True to make the img RGB # crop=False
blob = cv2.dnn.blobFromImage(img, 1/255.0, (416,416), swapRB=True)

# set the input to the network
net.setInput(blob)

# get output layers
output_layers = net.getUnconnectedOutLayersNames()
# output_layers = getOutputsNames(net)

# get the predictions = forward pass
predictions = net.forward(output_layers)

# post process
postprocess(img, predictions)

plt.figure(figsize=(15,10))
plt.imshow(img)
```
