```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
image_original = cv2.cvtColor(cv2.imread('images/dog.jpg'), cv2.COLOR_BGR2RGB)
image = image_original.copy()

points = np.array([[124,110],[124,135],[180,135],[180,110]], np.int32).reshape((-1,1,2))
cv2.fillPoly(image, [points], (255,0,0))
cv2.putText(image, 'bicycle', (125,125), cv2.FONT_HERSHEY_SIMPLEX, .5, (0,0,0), 1)
points2 = np.array([[125,135],[125,420],[565,420],[565,135]], np.int32).reshape((-1,1,2))
cv2.polylines(image, [points2], True, (255,0,0), 2)

points3 = np.array([[129,195],[129,220],[160,220],[160,195]], np.int32).reshape((-1,1,2))
cv2.fillPoly(image, [points3], (255,255,0))
cv2.putText(image, 'dog', (130,210), cv2.FONT_HERSHEY_SIMPLEX, .5, (0,0,0), 1)
points4 = np.array([[130,220],[130,540],[310,540],[310,220]], np.int32).reshape((-1,1,2))
cv2.polylines(image, [points4], True, (255,255,0), 2)

points5 = np.array([[479,55],[479,80],[524,80],[524,55]], np.int32).reshape((-1,1,2))
cv2.fillPoly(image, [points5], (0,255,0))
cv2.putText(image, 'truck', (480,70), cv2.FONT_HERSHEY_SIMPLEX, .5, (0,0,0), 1)
points6 = np.array([[480,80],[480,170],[690,170],[690,80]], np.int32).reshape((-1,1,2))
cv2.polylines(image, [points6], True, (0,255,0), 2)

plt.figure(figsize=(20,15))
plt.imshow(image, cmap='gray')
# cv2.imshow('Image', image)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
# cv2.waitKey(1)
```
