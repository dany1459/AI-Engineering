```python
# Strive2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

```python
img = cv2.imread('images/tiger1.jpg', cv2.IMREAD_GRAYSCALE)
ret, threshold = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

plt.figure(figsize=(15,10))
plt.imshow(threshold, cmap='gray')
```

## Dilation

```python
kernel = np.ones((3,3), np.uint8)
print(kernel)

dilation = cv2.dilate(threshold, kernel, iterations=2)

plt.figure(figsize=(15,10))
plt.imshow(dilation, cmap='gray')
```

```python
kernel_cross = cv2.getStructuringElement(cv2.MORPH_CROSS, (7,7))
print(kernel_cross)

dilation2 = cv2.dilate(threshold, kernel_cross, iterations=2)

plt.figure(figsize=(15,10))
plt.imshow(dilation2, cmap='gray')
```

```python
kernel_ellipse = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (7,7))
print(kernel_ellipse)

dilation3 = cv2.dilate(threshold, kernel_ellipse, iterations=2)

plt.figure(figsize=(15,10))
plt.imshow(dilation3, cmap='gray')
```

## Erosion

```python
erosion = cv2.erode(threshold, kernel, iterations=2)

plt.figure(figsize=(15,10))
plt.imshow(erosion, cmap='gray')
```

# Opening

```python
# erosion followed by dilation
opening = cv2.morphologyEx(threshold, cv2.MORPH_OPEN, kernel, iterations=2)

plt.figure(figsize=(15,10))
plt.imshow(opening, cmap='gray')
```

## Closing

```python
# dilation followed by erosion
closing = cv2.morphologyEx(threshold, cv2.MORPH_CLOSE, kernel, iterations=2)

plt.figure(figsize=(15,10))
plt.imshow(closing, cmap='gray')
```
