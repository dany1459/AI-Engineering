```python
square = np.zeros([500,500], np.uint8)

points = np.array([[75,300],[75,50],[325,50],[325,300]], np.int32).reshape((-1,1,2))
cv2.fillPoly(square, [points], (255,255,255))

plt.figure(figsize=(15,10))
plt.imshow(square, cmap='gray')
```

```python
circle = np.zeros([500,500], np.uint8)

cv2.circle(circle, (325,300), 150, (255,255,255), cv2.FILLED)

plt.figure(figsize=(15,10))
plt.imshow(circle, cmap='gray')
```

#### Bitwise functions = we compare 2 pixels, 1 from each image from the same position

```python
# shows pixels with a value > 0 only if both pixels > 0 (where they overlap) -> 
# output value = 'ON'
# otherwise, output value = 'OFF' = 0
bitwiseAND = cv2.bitwise_and(circle, square)

plt.figure(figsize=(15,10))
plt.imshow(bitwiseAND, cmap='gray')
```

```python
# shows pixels with a value > 0 if either of the 2 pixels is > 0
bitwiseOR = cv2.bitwise_or(circle, square)

plt.figure(figsize=(15,10))
plt.imshow(bitwiseOR, cmap='gray')
```

```python
# shows pixels with a value > 0 only if 1 of the 2 pixels is > 0 but NOT BOTH
# if both pixels > 0 -> output value = OFF = 0
bitwiseXOR = cv2.bitwise_xor(circle, square)

plt.figure(figsize=(15,10))
plt.imshow(bitwiseXOR, cmap='gray')
```

```python
# inverts the ON and OFF pixels in an image
# all pixels > 0 are set to 0, 
# all pixels = 0 are set to 255
bitwiseNOT = cv2.bitwise_not(square)

plt.figure(figsize=(15,10))
plt.imshow(bitwiseNOT, cmap='gray')
```
