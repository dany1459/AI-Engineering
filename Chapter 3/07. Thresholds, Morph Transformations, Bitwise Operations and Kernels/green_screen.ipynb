```python
# Strive2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

```python
elmer = cv2.cvtColor(cv2.imread('images/elmer_fudd.jpg'), cv2.COLOR_BGR2RGB)
tiger = cv2.cvtColor(cv2.imread('images/tiger1.jpg'), cv2.COLOR_BGR2RGB)
tiger = cv2.resize(tiger, (1280,720))
print(elmer.shape, tiger.shape)

plt.figure(figsize=(15,10))
plt.axis('off')
plt.imshow(elmer)
```

```python
lower_bounds = (0,80,0)
upper_bounds = (120,255,120)
mask = cv2.inRange(elmer, lower_bounds, upper_bounds)
elmer_copy = elmer.copy()
elmer_copy[mask != 0] = [0,0,0]

plt.figure(figsize=(15,10))
plt.axis('off')
plt.imshow(elmer_copy)
```

```python
plt.figure(figsize=(15,10))
plt.axis('off')
plt.imshow(mask, cmap='gray')
```

```python
tiger_copy = tiger.copy()
tiger_copy[mask == 0] = [0,0,0]

plt.figure(figsize=(15,10))
plt.axis('off')
plt.imshow(tiger_copy)
```

```python
elmer_tiger = tiger_copy + elmer_copy

plt.figure(figsize=(15,10))
plt.axis('off')
plt.imshow(elmer_tiger)
```
